## Python入门：以组合数计算为例

```python
# input: (n, m)
input_list = [(5, 3), (10, 2), (6, 0), (6, 6)]
```

### 方法一：定义函数，直接计算组合数C(n,m)

- 组合数计算公式：$C_n^m=\frac{n!}{(n-m)!m!}$

```python
def comb(n, m):
    ans = 1
    # TODO:直接计算组合数C(n,m)
    up = 1
    down1 = 1
    down2 = 1
    for i in range(1, n+1):
        up *= i
    for i in range(1, m+1):
        down1 *= i
    for i in range(1, n-m+1):
        down2 *= i
    ans = up // (down1 * down2)
    return ans

for n, m in input_list:
    ans = comb(n, m)
    print(f"n={n}, m={m}, ans={ans}")
```

### 方法二：定义函数，递推计算组合数

- 组合数计算公式：$C_i^j = C_{i-1}^{j-1} + C_{i-1}^j$，又称**帕斯卡法则**

- 实现上为**杨辉三角**，即如下形式的动态规划表

    | n \ m | 0 | 1 | 2 | 3 | 4 | 5 |
    | :--- | :--- | :--- | :--- | :--- | :--- | :--- |
    | **0** | **1** | | | | | |
    | **1** | 1 | 1 | | | | |
    | **2** | 1 | 2 | 1 | | | |
    | **3** | 1 | 3 | 3 | 1 | | |
    | **4** | 1 | 4 | 6 | 4 | 1 | |
    | **5** | 1 | 5 | 10 | 10 | 5 | 1 |

```python
def pascal_comb(n, m):
    c = [[0 for j in range(n + 1)] for i in range(n + 1)]   # nxn的零矩阵

    for i in range(n + 1):    # 列遍历
        # TODO:迭代推导组合数C(n,m)
        # 行遍历， 使用上一行的数据，初始有一个1
        for j in range(i+1):
            if j == 0:    # 处理左右边界
                c[i][j] = 1
            elif j == i:
                c[i][j] = 1
            else:
                c[i][j] = c[i-1][j-1] + c[i-1][j]     # 三角形

    return c[n][m]

for n, m in input_list:
    ans = pascal_comb(n, m)
    print(f"n={n}, m={m}, ans={ans}")
```

### 方法三：类封装+缓存表

```python
class Combinator:
    def __init__(self):
        # self.c 是一个二维列表，用于存储杨辉三角（组合数表）
        # 初始化第 0 层：C(0, 0) = 1
        self.c = [[1]]  # c[n][m] 对应组合数 C(n, m)

    def build(self, n):
        """
        根据当前缓存情况，动态增长杨辉三角至第 n 层。
        提示：利用递推公式 C(n, m) = C(n-1, m-1) + C(n-1, m)
        """
        # TODO: 使用 while 循环，从当前最后一层开始，逐层构建直到第 n 层
        current_layer = len(self.c)    # 第n层时的长度是n+1
        while current_layer <= n:
            current_list = [1]
            for i in range(1, current_layer):
                current_list.append(self.c[current_layer-1][i-1] + self.c[current_layer-1][i])
            current_list.append(1)
            self.c.append(current_list)
            current_layer += 1
            
        
    def get(self, n, m):
        self.build(n)
        return self.c[n][m]

cb = Combinator()
for n, m in input_list:
    ans = cb.get(n, m)
    print(f"n={n}, m={m}, ans={ans}")
```
