## Numpy入门

```python
import numpy as np
```

```python
# List
list_a = [1, 2, 3]
list_b = [3, 2, 1]
print(list_a, type(list_a))
print(list_b, type(list_b))

# 1D
np_a = np.array(list_a)
np_b = np.array(list_b)
print(np_a, type(np_a))
print(np_b, type(np_b))

# 2D
A = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
B = np.array([[3, 2, 1], [3, 2, 1], [3, 2, 1]])
print(A, type(A)) 
print(B, type(B))
```

###  List vs NumPy 差异

```python
# + 对照
print("== 加法 ==")
print("list_a + list_b =\n", list_a + list_b)
print("np_a + np_b =\n", np_a + np_b)
print("A + B =\n", A + B)

# * 对照
print("== 乘法 ==")
print("list_a * 2 =\n", list_a * 2)
print("np_a * 2 =\n", np_a * 2)
print("A * B =\n", A * B)
```

#### 向量计算

```python
u = np.array([1, 2, 3])
v = np.array([3, 2, 1])
```

```python
# TODO: 两种方式（符号@、numpy api）计算向量 u 与 v 的“内积”
# - 内积：对应位置元素相乘后求和，结果是一个标量
inner_uv_1 = u@v
inner_uv_2 = np.inner(u, v)
print("Inner product:", inner_uv_1, inner_uv_2)


# TODO: 使用numpy api计算向量 u 与 v 的“外积”
# - 外积：得到一个矩阵，矩阵第 (i,j) 项 = u[i] * v[j]
outer_uv = np.outer(u, v)
print("Outer product: ", outer_uv)


# TODO: 两种方式（numpy api / 自行等价实现）计算向量的 L2 范数（欧几里得长度）
# - L2 范数：sqrt(sum(u[i]^2))
norm_u_1 = np.linalg.norm(u - v)
norm_u_2 = np.sqrt(np.sum(np.square(u - v)))
print("||u||:", norm_u_1, norm_u_2)

# 将 NumPy 数组转换回 Python list 形式，并输出
print("outer_uv as python list:", outer_uv.tolist())
print("Type: ", type(outer_uv.tolist()))
      
```

#### 矩阵计算

```python
A = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
B = np.array([[3, 2, 1], [3, 2, 1], [3, 2, 1]])
```

```python
# TODO: 输出矩阵 A 与 B 的 shape（形状），用于确认矩阵维度是否匹配
print("A's shape =", A, "B.shape =", B)

# TODO: 两种方式（符号+、numpy api）计算矩阵加法 A + B（逐元素相加）
add_AB_1 = A + B
add_AB_2 = np.add(A, B)
print("A + B:\n", add_AB_1)
print("A + B:\n", add_AB_2)
print()

# TODO: 两种方式（符号*、numpy api）计算 element-wise 乘法 A * B（逐元素相乘）
elem_mul_AB_1 = A * B
elem_mul_AB_2 = np.multiply(A, B)
print("A * B (element-wise):\n", elem_mul_AB_1)
print("A * B (element-wise):\n", elem_mul_AB_2)
print()

# TODO: 两种方式（符号@、numpy api）计算矩阵乘法（matrix product）
# - 矩阵乘法：结果第 (i,j) 项 = A 第 i 行 与 B 第 j 列 的内积
mat_mul_AB_1 = A@B
mat_mul_AB_2 = np.dot(A, B)
print("matrix product:\n", mat_mul_AB_1)
print("matrix product:\n", mat_mul_AB_2)
print()

# TODO: 两种方式（符号.T、numpy api）计算矩阵转置（行列互换）
AT_1 = A.T
AT_2 = np.transpose(A)
print("A.T:\n", AT_1)
print("A.T:\n", AT_2)
```

## 矩阵进阶运算

```python
X = np.array([[1, 2, 4],[5, 4, 7],[9, 0, 8]])
b = np.array([10, 20, 30]) 
```

```python
print("X.shape =", X.shape, "b.shape =", b.shape)

# TODO: 两种方式（符号+、numpy api）计算矩阵 X 与向量 b 的加法 X + b，并观察当维度不匹配时，NumPy 是如何计算的
Y_1 = X + b
Y_2 = np.add(X, b)
print("\nY = X + b:\n", Y_1)
print("\nY = X + b:\n", Y_2)

# TODO: 使用 numpy api 计算结果每一「行」的和 row_sum，并保持二维矩阵的形状
row_sum = X.sum(axis=1, keepdims=True)
print("\nrow_sum:\n", row_sum, "\nshape:", row_sum.shape)

# TODO: 两种方式（符号/、numpy api）做“按行归一化”，使得每行和为 1
Y_norm_1 = X / row_sum
Y_norm_2 = np.divide(X, row_sum)
print("\nY_norm :\n", Y_norm_1)
print("\nY_norm :\n", Y_norm_2)
```

## 矩阵数据处理

```python
X = np.array([[1, 2, 4],[5, 4, 7],[9, 0, 8]])

# TODO: 切片（slicing）取子矩阵：取 X 的前两列。两种方式：直接索引 / numpy api
sub1_1 = X[:2:]
sub1_2 = X[slice(2)]
print("\n Slicing =\n", sub1_1)
print("\n Slicing =\n", sub1_2)

# TODO: 计算每「行」均值，两种方式：ndarray 方法 / numpy api
row_mean_1 = X.mean(axis=1)
row_mean_2 = np.mean(X, axis=1)
print("\nmean =", row_mean_1)
print("\nmean =", row_mean_2)

# TODO: 计算每「列」求和，两种方式：ndarray 方法 / numpy api
col_sum_1 = X.mean(axis=0)
col_sum_2 = np.mean(X, axis=0)
print("\nsum =", col_sum_1)
print("\nsum =", col_sum_2)

# TODO: 计算每「行」最大值的索引，两种方式：ndarray 方法 / numpy api
row_argmax_1 = X.argmax(axis=1)
row_argmax_2 = np.argmax(X, axis=1)
print("\nargmax =", row_argmax_1)
print("\nargmax =", row_argmax_2)
```
